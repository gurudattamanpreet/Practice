```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.runnables import RunnableLambda
from langchain_core.prompts import ChatPromptTemplate
from langgraph.graph import StateGraph,END
```

```python
llm=ChatGoogleGenerativeAI(model='models/gemini-1.5-flash',temperature=0.3)
```

```python
prompt=ChatPromptTemplate.from_template('user asked:{question}')
```

```python
def start_node(state):
    question = state["question"]
    response = llm.invoke(prompt.format_messages(question=question))
    return {"question": question, "response": response.content}
```

    | Concept         | Explanation                                            |
    | --------------- | ------------------------------------------------------ |
    | `TypedDict`     | Type-safe dictionary banata hai                        |
    | `ChatState`     | LangGraph ko batata hai tumhara data structure kya hai |
    | `question: str` | Dict me string value hone wali hai                     |
    | Kya fayda?      | Errors kam, workflow safe, type hints ka support       |


```python
from typing import TypedDict

class ChatState(TypedDict):
    question: str
    response: str
```

```python
builder = StateGraph(ChatState)
```

```python
builder.add_node("ask_llm", start_node)

builder.set_entry_point("ask_llm")

builder.set_finish_point("ask_llm")
```

```python
graph = builder.compile()
```

```python
user_question = input("Aap kya puchhna chahenge? ")

result = graph.invoke({"question": user_question})

print("🤖 Chatbot:", result["response"])
```
