```python
from sklearn.pipeline import Pipeline as Skpipeline
from imblearn.pipeline import Pipeline as Imbpipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report
from imblearn.over_sampling import SMOTE
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score, StratifiedKFold
```

    2. MLOps & Production Perspective
    Manual Approach Problems: smote outside pipeline

    ❌ Multiple steps to manage
    ❌ Data leakage risk
    ❌ Hard to deploy
    ❌ Difficult to version control
    ❌ Error-prone in production

    Pipeline Approach Benefits: smote inside pipeline
    Industry Approach:

    ✅ Single object to deploy
    ✅ No data leakage (automatic train/test split handling)
    ✅ Easy to serialize (joblib.dump)
    ✅ Reproducible
    ✅ CI/CD friendly

```python
num_cols = X_train.select_dtypes('number').columns.tolist()
cat_cols = X_train.select_dtypes('object').columns.tolist()

num_pipeline = Skpipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

cat_pipeline = Skpipeline([
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('encode', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer([
    ('num', num_pipeline, num_cols),
    ('cat', cat_pipeline, cat_cols)
], remainder='passthrough')
```

```python
pipe=Imbpipeline([
    ('preprocessor',preprocessor),
    ('pca',PCA(n_components=10)),
    ('smote',SMOTE(sampling_strategy=0.1)),
    ('model',LogisticRegression())])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_test_pred = pipe.predict(X_val)
print(classification_report(y_val, y_test_pred))
```

```python
accuracy_score(y_val,y_test_pred)
```

```python
cv_scores=cross_val_score(pipe, X_train, y_train,
                           cv=StratifiedKFold(n_splits=5),
                           scoring='accuracy')
```

```python
print(f"CV Mean: {cv_scores.mean():.3f}")
print(f"CV Std: {cv_scores.std():.3f}")  # High std = overfitting sign
```

```python
for classifier in classifiers:
    pipeline = Skpipeline(steps = [
               ('preprocessor', preprocessor)
              ,('classifier',classifier)
           ])
    model = pipe.fit(X_train, y_train)
    predictions = model.predict(X_val)
    print(classifier)
    print(classification_report(y_val, predictions))
```
