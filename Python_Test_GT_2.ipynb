```python
# Ek string ka first letter capital karo.
a='python is a programming language'
print(a.capitalize())
```

```python
# Ek string ko lowercase me convert karo using .casefold().
a='PYTHON IS PROGRAMMING LANGUAGE'
print(a.casefold())
```

```python
# String ko 20 width ke beech center karo using *.
a='python'
print(a.center(20,'*'))
```

```python
# String me kisi word ki occurrence count karo.
a='python is a programming language python'
print(a.count('python'))
```

```python
# String ko UTF-8 me encode karo.
a='python is a programming language python'
print(a.encode())
```

```python
# Check karo string kisi word se end hoti hai ya nahi.
a='python is a programming language python2'
if a[-1].isalpha():
  print('ends with word')
else:
  print('ends with something else')
```

```python
# String me \t ho to use space me badlo using .expandtabs().
a='python is a programming \t language python'
print(a.expandtabs())
```
