```python
# Ek string me vowels ki total count karo.
a='python is a programming language'
vowels='a','e','i','o','u'
sum(1 for x in a.split() if x[0].lower() in vowels)
```

```python
s1 = "python is a programming language."
def count_vowels(s1):
    vowels = "aeiouAEIOU"
    count = 0
    for char in s1:
        if char in vowels:
            count += 1
    return count

print(count_vowels(s1))
```
