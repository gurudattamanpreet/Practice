```python
# Write a program to merge two dictionaries.
a={'a':23,'b':34,'c':12,'d':21,'e':65}
b={'a':21,'g':32,'h':52,'i':61,'j':90}
print({x: a.get(x,0) + b.get(x,0) for x in set(a).union(set(b))})
```
