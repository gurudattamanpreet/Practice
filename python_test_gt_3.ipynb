```python
# Do dicts ko merge karo, aur common key ki value update ho.
dr={'a':40,'b':50,'c':60,'d':10}
nr={'e':10,'a':90,'r':30,'b':20}
{x: dr.get(x,0) + nr.get(x,0) for x in set(dr).intersection(set(nr))}
```
