```python
# zip() ka use karke 2 list ko combine karo and result list of tuples ho
a=[2,3,4,5,6,7,8]
b=[8,7,6,5,4,3,2]
print(list(zip(a,b)))
```

```python
# any() aur all() ka use karke ek boolean list evaluate karo.
boolen=[True,False,True,True,True]
print(any(boolen))
# any() check karta hai: kya list mein koi ek bhi True hai? Agar haan, to True return karta hai.
print(all(boolen))
```

```python
# Generator expression banao jo 1 to 50 tak ke even numbers ka sum de.
sum(x for x in range(51) if x%2==0)
```
