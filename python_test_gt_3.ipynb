```python
# zip() ka use karke 2 list ko combine karo and result list of tuples ho
a=[2,3,4,5,6,7,8]
b=[8,7,6,5,4,3,2]
print(list(zip(a,b)))
```

```python
# any() aur all() ka use karke ek boolean list evaluate karo.
boolen=[True,False,True,True,True]
print(any(boolen))
# any() check karta hai: kya list mein koi ek bhi True hai? Agar haan, to True return karta hai.
print(all(boolen))
```

```python
# Generator expression banao jo 1 to 50 tak ke even numbers ka sum de.
sum(x for x in range(51) if x%2==0)
```

```python
# Ek sentence lo, unique words count karo (case-insensitive).
a='Python is a rogramming language python is a'
sum(1 for x in set(a.lower().split()) if a.lower().split().count(x)<2)
```

```python
# setdefault() ka use karo new key add karne ke liye.
a={'a':40,'b':50}
a.setdefault('c',60)
print(a)
```

```python
# Ek dict se last inserted item hatao.
a={'a':40,'b':50,'c':60,'d':10}
a.popitem()
print(a)
```

```python
# dict.get() ka use karo safe access ke liye.
a={'a':40,'b':50,'c':60,'d':10}
a.get('d')
```

```python
# Do dicts ko merge karo, aur common key ki value update ho.
dr={'a':40,'b':50,'c':60,'d':10}
nr={'e':10,'a':90,'r':30,'b':20}
{x: dr.get(x,0) + nr.get(x,0) for x in set(dr).intersection(set(nr))}
```

```python
# Top 2 students nikaalo based on average marks in dict.
a={'name':['david','alex','micheal','grofer','gilchrist'],'marks':[90,49,45,76]}

d=sum(a['marks'])/len(a['marks'])

sorted([x for x in a['marks'] if x>d], reverse=True)[:3]
```
