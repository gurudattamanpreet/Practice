    PromptTemplate ek tool hai LangChain mein jisse hum prompts ko smart aur dynamic bana sakte hain.
    Isme placeholder variables hote hain, jaise {resume_text}
    Jab actual resume input aata hai, toh wo placeholders fill ho jaate hain – jese Google Form bharna.
    Phir AI ko final prompt bhej diya jaata hai to process and respond.

```python
template = """
You are a resume summarizer AI agent.

Given the resume below, summarize it in 5-6 lines.
Then extract key technical skills.
Then suggest 2 job roles that might fit this profile.

Resume:
{resume_text}

Respond in the following format:
- Summary:
- Key Skills:
- Suitable Job Roles:
"""

prompt = PromptTemplate(input_variables=["resume_text"], template=template)
# llm = ChatGoogleGenerativeAI(model="gemini-pro", temperature=0.3)
llm = ChatGoogleGenerativeAI(model="models/gemini-1.5-flash", temperature=0.3)
chain = prompt | llm
```

    Iska matlab:
      1.	{resume_text} fill hoga template mein
      2.	Prompt banega
      3.	Prompt AI model ko jayega
      4.	AI response dega: summary, skills, job roles

    llm.invoke(prompt_text) >>  AI ko prompt do, output lo
    chain.invoke({...}) >> Dono step ek saath karo (chain bana ke)


```python
resume_input = {"resume_text": resume_text}

response = chain.invoke(resume_input)
print(response.content)
```
