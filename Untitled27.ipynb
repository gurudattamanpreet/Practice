    Yeh function input lega ek dictionary jisko “state” bol rahe hain.
    LangGraph me “state” ek dictionary hoti hai jisme workflow ka input/output data hota hai.

    state dictionary me se "resume_text" key ka value nikal rahi hai.
  

    invoke() ka matlab: chain ko ek input do, aur output lo.


    LangChain me RunnableLambda ek wrapper hai
    jo kisi bhi Python function ko LangGraph node me convert karta hai.

```python
def summarize_step(state):
    input_text = state['resume_text']
    response = chain.invoke({"resume_text": input_text})
    return {
        "summary": response.content,
        "resume_text": input_text
    }
```

    📦 Graph node = ek step ya task in your pipeline
    🧠 Python function = normal kaam karne wala code
    🔁 RunnableLambda = “Ye function ab graph me ek valid step ban gaya”

    LangGraph me state ek dictionary hoti hai jo har step ke input/output ko track karti hai.
    Jab tu ek node (jaise summarize_node) execute karta hai, wo state update karta hai.

```python
summarize_node = RunnableLambda(summarize_step)
```

```python
graph = StateGraph(dict)
graph.add_node("summarize", summarize_node)
graph.set_entry_point("summarize")
graph.set_finish_point("summarize")
```

```python
template = """
You are a resume summarizer AI agent.

Given the resume below, summarize it in 5-6 lines.
Then extract key technical skills.
Then suggest 2 job roles that might fit this profile.

Resume:
{resume_text}

Respond in the following format:
- Summary:
- Key Skills:
- Suitable Job Roles:
"""

prompt = ChatPromptTemplate.from_template(template)
llm = ChatGoogleGenerativeAI(model="models/gemini-1.5-flash", temperature=0.3)
chain = prompt | llm
```

    app = graph.compile()
      •	graph abhi tak ek blueprint tha — jaise ki ek pipeline ka design
      •	.compile() usko final working machine bana deta hai

```python
app = graph.compile()

final_output = app.invoke({"resume_text": resume_text})

print(final_output["summary"])
```
