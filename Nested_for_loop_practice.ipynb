```python
# # Nested for loop ka use karke 1 se 3 tak ke numbers ke saath 1 se 2 tak ke numbers ka har possible pair print karo.

# output:

# (1, 1)
# (1, 2)
# (2, 1)
# (2, 2)
# (3, 1)
# (3, 2)


for i in range(1,4):
  for j in range(1,3):
    print(i,j)
```

Logic:

| i | j | Output     |
|---|---|------------|
| 1 | 1 | (1, 1)     |
| 1 | 2 | (1, 2)     |
| 2 | 1 | (2, 1)     |
| 2 | 2 | (2, 2)     |
| 3 | 1 | (3, 1)     |
| 3 | 2 | (3, 2)     |

Carry Forward Variable Rule

	•	Jab loop ke andar koi value yaad rakhni ho toh use loop ke bahar define karo

Dynamic Row Lengths

	•	Inner loop ka range outer loop se depend kare
	•	For triangle shapes, range(i), range(1, i+1) etc. use hote hain

 Inner loop can use outer loop’s value

	•	You can use i inside the inner loop to control its behavior


Pattern Planning Rule (Important!)

Har pattern se pehle ye 3 sawal socho:

	1.	Kitni rows chahiye? → outer loop
	2.	Har row me kya print hoga? → inner loop
	3.	Kya value har row me badal rahi hai ya carry forward ho rahi hai? → variables ka planning



```python
#     A
#    A B
#   A B C
#  A B C D
# A B C D E


for i in range(1,6):
  for s in range(5 - i):
    print(' ', end='') #pehle 5 ka space dega, phir 4 ka space dega & continues till 5-i
  for j in range(65,65+i):
    print(chr(j),end=' ')
  print()
```

```python
# A B C D E
#  A B C D
#   A B C
#    A B
#     A

for i in range(5):
  for s in range(i):
    print(' ',end='')
  for j in range(65,70-i):
    print(chr(j),end=' ')
  print()


# 65,65+i  seedha pyramid
# 65, 70-i  ulta pyramid

# Jab pattern shrink ho raha ho → range(..., end - i)
# Jab pattern grow ho raha ho → range(..., start + i)
```
