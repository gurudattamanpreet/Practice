Answer:

append() ki responsibility Python khud le raha hai jab tu square brackets [x for ...] likhta hai.
Yeh Python ka shortcut hai jo automatically list banata hai aur usme har x ko daal deta hai.

Python ka Rule:

You can loop through:
	•	list ([1, 2, 3])
	•	string ("ABC")
	•	tuple ((1, 2))

But you CANNOT loop through:
	•	int, float, etc.

```python
# tricky

# Group Numbers into ‘even’ and ‘odd’
# expected output:
# {
#   'even': [2, 4, 6],
#   'odd': [1, 3, 5]
# }

a = [5,8,4,9,1]
r={'even':[],'odd':[]}
for i in a:
  if i%2==0:
    r['even'].append(i)
  else:
    r['odd'].append(i)
print(r)
```

```python
# Group words by their length

# output:
# {
#   2: ['hi', 'go'],
#   3: ['sun', 'bye'],
#   5: ['world'],
#   6: ['python']
# }

a = ['hi', 'sun', 'world', 'bye', 'python', 'go']

# {key: [items matching key] for key in set}
print({i: [x for x in a if len(x) == i] for i in set(len(x) for x in a)})

# isne words ki length match krne k liye for x in a lagaya shuru me
# uske baad saare words ki length ka set banane k liye bhi x in a lagaya last me
```

```python
a = ['hi', 'sun', 'world', 'bye', 'python', 'go']

r = {}
for x in a:
    length = len(x)
    if length not in r:
        r[length] = []
    r[length].append(x)

print(r)
```

Step 1: Lengths of words in a
------------------------------------
Word      Length
----------------

'python'     6

Step 2: Unique lengths → using set
------------------------------------
set(len(x) for x in a) = {2, 3, 5, 6}

------------------------------------
First i = 2
    → [x for x in a if len(x) == 2]

Second i = 3
    → [x for x in a if len(x) == 3]
    → ['sun', 'bye']

Third i = 5
    → [x for x in a if len(x) == 5]
    → ['world']

Fourth i = 6
    → [x for x in a if len(x) == 6]
    → ['python']
