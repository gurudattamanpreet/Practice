```python
try:
  a=int(input('enter 1st number: '))
  b=int(input('enter 2nd number: '))
  print(a+b)
  print(a/b)
except ValueError:
  print('int dena h value nhi')
except ZeroDivisionError:
  print('zero se divide nhi kr skta')
else:
  print('nothing')
finally:
  print('har haal me chalega')
```

```python
try:
  age = int(input("Apni age batao: "))
  if age <18:
    raise
except:
  print('not eligible for vote')
```

```python
try:
    age = int(input("apni age batao: "))

    if age < 0:
        raise ValueError("age negative nhi ho skti")

    if age < 18:
        raise ValueError("voting ke liye 18 saal ka hona zaroori hai.")

    else:
      print("aap vote dene ke liye eligible ho.")

except ValueError as e:
    print(e)
```
