```python
class institute:
  def __init__(self,name,age,marks,contact):
    self.nama=name
    self.age=age
    self.marks=marks
    self.contact=contact
    print('institute has limited seats')
```

```python
s3=institute('nikita',21,100,1234567890)
```

```python
class student:
  school_name='python school'

  def __init__(self,name,age):
    self.name=name
    self.age=age

  def info(self):
    print(self)
    print(self.name)
    print(self.age)
    print(self.school_name)


  @classmethod
  def info2(cls):
    # print(cls)
    print(cls.name)
    print(cls.age)
    print(cls.school_name)
```

```python
s2=student('sjd',78)
```
