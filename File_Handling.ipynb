```python
with open('mydata.txt','w') as f:
  lines=f.write('python is awesome')
  print(lines)
```

```python
f=open('mydata.txt','r')
f.read()
```

```python
lines=['python\n','is\n','awesome\n']
with open('mydata.txt','w') as f:
    f.writelines(lines)
```

```python
with open('mydata.txt','r') as f:
  for i in f:
    print(i.strip())
```
